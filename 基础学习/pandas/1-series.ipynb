```python
import numpy as np
import pandas as pd
```

series是pandas中用于存储一维数据的对象，可以由列表、字典或numpy数组创建。

```python
# 一、创建series
## 1、用列表创建
```

```python
s1 = pd.Series([1, 2, 3, 4])
print(s1)

print('s1.values: ', s1.values)

print('s1.index: ', s1.index)

# 用numpy数组创建
s2 = pd.Series(np.arange(4))
print('s2: ', s2)

```

## 2、用字典创建

```python
s1 = pd.Series({'a': 1, 'b': 2, 'c': 3})
print(s1)
```

## 3、自定义索引

```python
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(s1)
print('s1["a"]: ', s1['a'])
print('s1[["a", "b"]]: ', s1[['a', 'b']])
print('s1[0]: ', s1[0])
print('s1[0:2]: ', s1[0:2])
print('s1[s1>2]: ', s1[s1>2])

```

# 二、将Series转换成其他类型
## 1、将Series转换成dict

```python
s1 = pd.Series({'a': 1, 'b': 2, 'c': 3})
print('s1: ', s1)

s1.to_dict()
print('s1.to_dict(): ', s1.to_dict())

```
