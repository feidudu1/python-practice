```python
import numpy as np
import pandas as pd
```

是pandas中的二维数据结构，可以通过列表、字典、series等方式创建。
# 一、创建dataFrame

```python
## 1、list转dataframe
```

```python
df = pd.DataFrame([
  [1, 2],
  [3, 4],
])
print('df: ', df)


```

## 2、dict转dataframe

```python
data = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}
df = pd.DataFrame(data)
print('df: ', df)
```

## 3、series转dataframe

```python
series = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
df = pd.DataFrame(series)
print('df: ', df)

df2 = pd.DataFrame({'col1': series})
print('df2: ', df2)

df3 = pd.DataFrame({'col1': series, 'col2': series})
print('df3: ', df3)
```

## 4、numpy数组转dataframe

```python
array = np.array([[1, 2, 3], [4, 5, 6]])
df = pd.DataFrame(array)
print('df: ', df)
```

## 5、json数据转成dataFrame

```python
my_json = '{"col1": [1, 2, 3], "col2": [4, 5, 6]}'
df = pd.read_json(my_json)
print('df: ', df)
```

## 6、其他

```python
df = pd.DataFrame([1,2,3,4], index=['a', 'b', 'c', 'd'], columns=['col1'])
print('df: ', df, '\n')

df = pd.DataFrame(np.arange(12).reshape(3,4))
print('df: ', df, '\n')
```

# 二、获取索引值和列名
## 1、基础

```python
df = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6]})
print('df: ', df, '\n')

print('df.index: ', df.index, '\n')
print('df.columns: ', df.columns, '\n')
print('df.columns.values: ', df.columns.values, '\n')
print('df.values: ', df.values, '\n')

print('df.head(1): ', df.head(1), '\n')
print('df.tail(1): ', df.tail(1), '\n')

print('df.describe(): ', df.describe(), '\n')

```

## 2、通过index获取/iloc

```python
print('df.具体值: ', df[['col1']], '\n')
print('df.具体值.type: ', type(df[['col1']]), '\n')

print('df.具体值2: ', df['col1'], '\n')
print('df.具体值2.type: ', type(df['col1']), '\n')

print('df.具体值3: ', df['col1'][0], '\n')
print('df.具体值3.type: ', type(df['col1'][0]), '\n')

print('df.具体值4: ', df[['col1']][1:], '\n')
print('df.具体值4.type: ', type(df['col1'][1:]), '\n')

print('df.具体值5: ', df[['col1']].iloc[1:], '\n')
print('df.具体值5.type: ', type(df['col1'].iloc[1:]), '\n')




```

## 3、通过label获取/loc

```python
df = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6], 'col3': [7, 8, 9]})
print('df: ', df, '\n')

print('df.loc[0]: ', df.loc[0], '\n')
print('df.loc[0].type: ', type(df.loc[0]), '\n')

print(df[['col1']], '\n')
print(df[['col1']].loc[0], '\n')
print(df[['col1', 'col2']].columns.values, '\n')
```

```python
print(df['col1'], '\n')
print(type(df['col1']), '\n')
print(df['col1'].loc[0], '\n')
```
